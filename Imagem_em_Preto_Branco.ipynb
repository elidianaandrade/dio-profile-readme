```python
from PIL import Image

def converter_para_cinza(imagem_colorida, imagem_saida):
    # Abrir a imagem
    imagem_colorida = Image.open(imagem_colorida)
    # Imagem para RGB
    imagem_conver_RGB = imagem_colorida.convert("RGB")

    # Criar uma imagem para servir de base
    imagem_base = Image.new("RGB", imagem_conver_RGB.size)

    # Declarar a largura para percorrer a imagem
    largura, altura = imagem_conver_RGB.size

    # Percorrer a imagem para transformar a imagem colorida em escala de cinza
    for x in range(largura):
        for y in range(altura):
            r, g, b = imagem_conver_RGB.getpixel((x, y))

            cinza = int(0.299 * r + 0.587 * g + 0.114 * b)

            imagem_base.putpixel((x, y), (cinza, cinza, cinza))

    # Salva a imagem resultante
    imagem_base.save(imagem_saida)
    print(f"Imagem em tons de cinza salva como {imagem_saida}")


def converter_preto_branco(imagem_cinza, imagem_saida):
    # Abrir a imagem
    imagem_cinza = Image.open(imagem_cinza)
    # Converter para escala de cinza (se necessário)
    imagem_conver_RGB = imagem_cinza.convert("RGB")

    # Criar uma imagem para servir de base
    imagem_base = Image.new("RGB", imagem_conver_RGB.size)

    # Declarar a largura para percorrer a imagem
    largura, altura = imagem_conver_RGB.size

    # Percorrer a imagem para transformar em preto e branco
    for x in range(largura):
        for y in range(altura):
            r, g, b = imagem_conver_RGB.getpixel((x, y))
            if r < 128:
                imagem_base.putpixel((x, y), (0, 0, 0))  # Preto
            else:
                imagem_base.putpixel((x, y), (255, 255, 255))  # Branco

    # Salva a imagem resultante
    imagem_base.save(imagem_saida)
    print(f"Imagem em preto e branco salva como {imagem_saida}")


# Exemplo de uso
converter_para_cinza('imagem-colorida_natureza.jpg', 'imagem_cinza.jpg')
converter_preto_branco('imagem_cinza.jpg', 'imagem_preto_branco.jpg')
```
